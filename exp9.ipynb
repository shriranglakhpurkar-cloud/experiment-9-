```python
import numpy as np
a = np.array([1,10,20,30,40,50,500])
b = np.array([[1,2,3],[4,5,6]])
print(a)
print(b)
```

```python
#Dimension of Array
print(a.ndim)
print(b.ndim)

#Size oe shape or number of rows and columns
print(a.shape)
print(b.shape)

#Datatype
print(a.dtype)
print(b.dtype)

#In-Built functions using NumPy
print(np.zeros((2,3)))
print()
print(np.ones((3,3)))

print()
print(np.eye((3)))
print()

print(np.arange(1,10,2))
print(np.linspace(0,1,4))
print(b * 2)
print()
print(a + 5)
print(np.mean(a))
print(np.median(a))

print(np.max(a))
print(np.min(a))

print(np.sum(a))
```
